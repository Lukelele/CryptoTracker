```python
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import torch
import beacon
```

```python
plt.style.use("dark_background")
```

```python
# Replace 'LTC-USD' with the desired Yahoo Finance ticker symbol for Litecoin
symbol = 'LTC-GBP'
df = yf.download(symbol, start="2017-04-10", end=None)
Open = df["Open"]
High = df["High"]
Low = df["Low"]
Close = df["Close"]
Volume = df["Volume"]
df.head()
```

```python
fig, ax = plt.subplots(1, 2, figsize=[12, 4])
ax[0].plot(df["Close"])
ax[0].set_title("Closing Price")
ax[1].plot(df["Volume"])
ax[1].set_title("Volume")
```

```python
normalised_close = Close / Close.max()
normalised_volume = Volume / Volume.max()

plt.figure(figsize=(12, 5))
plt.plot(normalised_close, label="Close")
plt.plot(normalised_volume, label="Volume", c="white", alpha=0.5)
plt.legend()
```

```python
price_volume_ratio = normalised_volume / normalised_close
normalised_ratio = price_volume_ratio / price_volume_ratio.max()

plt.figure(figsize=(12, 5))
plt.plot(normalised_close, label="Close")
plt.plot(normalised_ratio, label="Volume to Price Ratio", alpha=0.5)
plt.legend()
```
